### **PURCHASE MODULE:**
1. USER CAN BUY A PRODUCT
2. DISCOUNT IS AVAILABLE FOR PRODUCTS
3. USER CAN APPLY COUPONS ON A MINIMUM AMOUNT OF PURCHASE
4. USER CAN RATE A PRODUCT
5. USER CAN SEE THE RATINGS OF A PRODUCTS
6. USER CAN SEE HOW MANY TIMES A PRODUCT IS SOLD
7. THE SALES FILE CAN BE USED TO SEE ALL THE PURCHASE HISTORY INCLUDING BILLING AMOUNT AND TIME

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

###Sales Dictionary

```python
sales= dict()
```

```python
print("We have 3 coupons available:")
print("On purchase of 1000 and above - Get flat 15% off - use PROMO CODE: SHOP15")
print("On purchase of 2500 and above - Get flat 25% off - use PROMO CODE: SHOP25")
print("On purchase of 5000 and above - Get flat 35% off - use PROMO CODE: SHOP35")

ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))
f=0
for id in records:
  if(id == ui_prod):
    if ui_quant > records[id]['qn']:
      print("We only have ",records[id]['qn']," ",records[ui_prod]['name'])
      f=1
      break
    print("You ordered: ", records[ui_prod]['name'])
    print("Quantity: ", ui_quant)
    print("Total purchases of the product are: ",records[ui_prod]['Times_purchased'])
    print("Price: ", records[ui_prod]['pr'])
    print("Discount: ", records[ui_prod]['discount'],"%")
    print("Billing Amount: ", records[ui_prod]['pr'] * ui_quant)
    total_amount=(records[ui_prod]['pr'] * ui_quant) - ((records[ui_prod]['pr'] * ui_quant)*(records[ui_prod]['discount'])/100)
    print("Billing Amount after discount: ", total_amount)

    records[ui_prod]['Times_purchased'] = records[ui_prod]['Times_purchased']+ui_quant
    records[ui_prod]['qn'] = records[ui_prod]['qn'] - ui_quant
    break
else:
  print("We don't have a product of this ID")
  f=1
if f!=1:
  if records[ui_prod]['pr'] * ui_quant > 5000:
    print("You have purchase of greater than 5000, do you want to use coupon SHOP35?")
    yesno = input("y/n: ")
    if yesno == 'y':
      total_amount=(records[ui_prod]['pr'] * ui_quant)-((records[ui_prod]['pr'] * ui_quant)*0.35)
  elif records[ui_prod]['pr'] * ui_quant > 2500:
    print("You have purchase of greater than 2500, do you want to use coupon SHOP25?")
    yesno = input("y/n: ")
    if yesno == 'y':
      total_amount=(records[ui_prod]['pr'] * ui_quant)-((records[ui_prod]['pr'] * ui_quant)*0.25)
  elif records[ui_prod]['pr'] * ui_quant > 1000:
    print("You have purchase of greater than 1500, do you want to use coupon SHOP15?")
    yesno = input("y/n: ")
    if yesno == 'y':
      total_amount=(records[ui_prod]['pr'] * ui_quant)-((records[ui_prod]['pr'] * ui_quant)*0.15)
  print("Your total bill: ", total_amount)
import time
sales[len(sales)+1]={'prod_id' : ui_prod, 'prod_name': records[ui_prod]['name'],'prod_qn':ui_quant, 'Total_amount':total_amount , 'Purchased_date':time.ctime()}
```

```python
sales
```

###Rate a Product

```python
print("Do you want to rate a product?")
rateyn = input("y/n: ")
if rateyn == 'y':
  id_rate  = str(input("Enter the product_Id: "))
  user_rating = int(input("Rate from 1 to 5: "))
  if records[id_rate]['ratings'] == 'Unrated':
    records[id_rate]['ratings']=0
  
  past_rate = records[id_rate]['ratings']
  times_rated = records[id_rate]['Times_rated']
  records[id_rate]['ratings'] = ((past_rate*times_rated)+user_rating)/(times_rated+1)
  records[id_rate]['Times_rated'] = times_rated+1;

  print("Avg Rating ",records[id_rate]['ratings'])
  print("Rated By ",records[id_rate]['Times_rated'])
```

###Update Inventory after purchase

```python
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
